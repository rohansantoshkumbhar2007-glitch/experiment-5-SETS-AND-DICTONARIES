```python
'''PROBLEM STATEMENT 3:One set contains students in the cricket club , another set contains students in the badminton club.
Find students who are:
 1) In both clubs
 2)In only one club

 Concept Used : INTERSECTION,SYMMETRIC DIFFERENCE'''

 # Define the sets
cricket = {"Virat", "Dhoni", "Rohit", "Hardik","Messi"}
football = {"Messi", "Ronaldo", "Neymar", "Mbappe"}

# 1) Students in both clubs (Intersection)
both_clubs = cricket.intersection(football)

# 2) Students in only one club (Symmetric Difference)
only_one_club = cricket.symmetric_difference(football)

# Print results
print("Students in both clubs:", both_clubs)
print("Students in only one club:", only_one_club)
```

```python
'''PROBLEM STATEMENT 4: Absent Students
given a set of all students and a set of present students, find the students who are absent'''

# Define the sets
all_students = {"Messi", "Ronaldo", "Neymar", "Mbappe",}
present_students = {"Messi","Neymar","Mbappe",}
absent_students = all_students.difference(present_students)

# Print the result
print("Absent Students:", absent_students)
```

```python
''' PROBLEM STATEMENT 5: Remove Invalid Entries.
A set contains valid course codes.
Remove a discontinues course codes from the set.
Concept Used: remove()/ discard()'''


# Set of valid course codes
courses = {"DCLD", "AMT", "TPCS", "CT"}

print("Original courses:", courses)

# Discontinued course
discontinued_course = "TPCS"

# Using remove()
courses.remove(discontinued_course)
print("After removing using remove():", courses)

# Adding it back for demonstration
courses.add("TPCS")

# Using discard()
courses.discard("TPCS")
print("After removing using discard():", courses)
```

```python
'''1) A dictionary stores product names and their prices. Write a program to update the price of a given product'''
# Dictionary with product names and prices
products = {
    "apple": 100,
    "banana": 40,
    "orange": 60
}

# Input from user
product_name = input("Enter product name: ")
new_price = float(input("Enter new price: "))

# Update price if product exists
if product_name in products:
    products[product_name] = new_price
    print(f"Updated price of {product_name}: {products[product_name]}")
else:
    print("Product not found!")

# Display updated dictionary
print("Updated Dictionary:")
print(products)
```

```python
'''2) A dictionary stores student names and their marks. Write a program to search for a students's marks using the name entry. If the student is not found, display a suitable message.'''
# Dictionary storing student names and their marks
students = {
    "Alice": 85,
    "Bob": 78,
    "Charlie": 92,
    "David": 88
}

# Take student name as input
name = input("Enter the student's name: ")

# Search for the student's marks
if name in students:
    print(f"{name}'s marks are: {students[name]}")
else:
    print("Student not found.")
```

```python
'''3) A dictionary stores usernames and passwords. Write a python program to validate user login.'''
# Dictionary storing usernames and passwords
users = {
    "alice": "alice123",
    "bob": "bob@2024",
    "charlie": "charlie_pwd"
}

# Take input from user
username = input("Enter username: ")
password = input("Enter password: ")

# Validate login
if username in users and users[username] == password:
    print("Login successful!")
else:
    print("Invalid username or password.")
```

```python
'''4)A dictonary stores student names and marks. Write a Python program to find the student with the highest marks.'''
# Dictionary storing student names and their marks
students = {
    "REHMAN": 85,
    "ASLAM": 92,
    "IQBAL": 88,
    "AJAY": 95
}

# Find the student with the highest marks
top_student = max(students, key=students.get)

print("Student with highest marks:", top_student)
print("Marks:", students[top_student])
```
